## Part 1: Data preparation

```python
# Toggle settings
#add_noise = True
add_noise = False

# Initialize lists for image collection
train_images = []
test_images = []

# Paths for training and test datasets
train_paths = [
    'PandasBears/Train/Pandas',
    'PandasBears/Train/Bears'
]
test_paths = [
    'PandasBears/Test/Pandas',
    'PandasBears/Test/Bears'
]

# Dataset paths
datasets = [train_paths, test_paths]

for i, dataset_paths in enumerate(datasets):
    # Determine if it's training or testing dataset
    datasetname = ['Train', 'Test'][i]
    images = train_images if i == 0 else test_images

    for folder_path in dataset_paths:
        # Print the name of the folder that is currently being processed
        print(folder_path, end=' ')

        # Initialize file count
        file_count = 0

        # Go through all files in the folder
        for filename in os.listdir(folder_path):
            # Find the files that are JPEGs
            if filename.endswith(".jpg") or filename.endswith(".jpeg"):
                # Add 1 to the file count
                file_count += 1

                # Construct full file path
                file_path = os.path.join(folder_path, filename)

                # Import image
                image = plt.imread(file_path, format='jpeg')

                # Convert to grayscale
                image = np.dot(image[..., :3], [0.2989, 0.5870, 0.1140])

                # Decrease image size by 50%
                image = image[::2, ::2]

                if add_noise:
                    # Add some noise
                    image = image + np.random.normal(scale=100, size=image.shape)

                # Add the new image to collection
                images.append(image)

        print('has {} images'.format(file_count))
```

```python
# look at 4 random bears
for i0, i in enumerate(np.random.randint(0, 500, size=4)):
    plt.subplot(2,2,1+i0)
    plt.imshow(train_images[i][::2,::2],cmap='Greys_r')
```

# Part 2: Singular value decomposition

```python
import numpy as np
from PIL import Image

# Resize images to 128x128 and ensure all images are grayscale
resized_images = [np.asarray(Image.fromarray(image).resize((128, 128)).convert('L')) for image in train_images]

# Confirm image size
print("Shape of resized image:", resized_images[0].shape)  # Expected: (128, 128)

# Convert each image to a 1D array and create a matrix where each row is a flattened image
flattened_images = np.vstack([image.ravel() for image in resized_images])

# Calculate the mean image from the training dataset
mean_image = np.average(flattened_images, axis=0)

# Center the data by subtracting the mean image from each flattened image
centered_data = flattened_images - mean_image
A = centered_data  # A will have the shape (samples, features), e.g., (500, 16384) for 500 images

# Apply Singular Value Decomposition (SVD) to the centered data
U, S, Vh = np.linalg.svd(A, full_matrices=False)
```

```python
# show the first four eigenbears
test_images_flat = [img.flatten() for img in test_images]  # Flatten each image
mean_image = np.mean(np.stack(test_images_flat), axis=0)  # Calculate mean from the training set or directly here if not done
centered_test_images = [img - mean_image for img in test_images_flat]

for i in range(4):
    plt.subplot(2, 2, i + 1)
    # Reshape the ith row of Vh to the original image dimensions
    #change it to be vh based on the transpose to fit the original dimensions
    eigenbear = Vh[i].reshape((128, 128))
    #force colormap to be the same for all four
    plt.imshow(eigenbear, cmap='Greys_r', vmin=-np.max(np.abs(Vh[:4, :])), vmax=np.max(np.abs(Vh[:4, :])))
    plt.colorbar()  # Display colorbar for scale
plt.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust space between plots for clarity
plt.show()
```

```python
# Clustering of panda bears and brown bears along the first and second principal component
# indices of pandas in the test set
indices_pandas = range(50)
indices_brownbears = range(50,100)

for i, indices in enumerate([indices_pandas, indices_brownbears]):
    # Get projections of data onto principal component 1
    p1 = [np.dot(Vh[0], centered_test_images[x]) for x in indices]  # Use the first row of Vh for PC1
    # Get projections of data onto principal component 2
    p2 = [np.dot(Vh[1], centered_test_images[x]) for x in indices]  # Use the second row of Vh for PC2
    plt.plot(p1, p2, marker='+x'[i], lw=0, label=['Pandas', 'Brown Bears'][i])

# Annotate axes
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')
# Add legend
plt.legend()
plt.show()
```

# Part 3: Nonparametric classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# construct response variable: Train set was created by appending 250 pandas
# and THEN 250 brown bears to the list of training images. We code pandas as
# '0' and brown bears as '1'.
y_train =  np.concatenate([np.zeros(250), np.ones(250)])

# Test set was created by appending 50 pandas and THEN 50 brown bears to the
# list of test images. We code pandas as '0' and brown bears as '1'.
y_test = np.concatenate([np.zeros(50), np.ones(50)])
```

```python
# Verify the shape of A
print("Shape of A before any operation:", A.shape)

# Ensure A is shaped (500, 16384), if not, transpose it correctly
if A.shape[0] != 500:
    A = A.T
    print("Transposed A for correct orientation, new shape:", A.shape)

errors_list = []

# Now, fit the model with A without transposing
print('   k\t|  # errors\t| misclassified bears')
print('--------------------------------------------')
for k in range(1, 16):
    # Fit KNN model
    modelKN = KNeighborsClassifier(n_neighbors=k).fit(A, y_train)

    # Use model to make predictions on the test set
    # Ensure test images are also shaped correctly before prediction
    predictions = [modelKN.predict([centered_test_images[i]]) for i in range(len(y_test))]

    # Detect misclassifications
    errors = np.abs(np.array(predictions).flatten() - y_test)  # Ensure predictions are correctly flattened
    for i in np.argwhere(errors).T[0]:
        errors_list.append(i)
    # Print results to table
    print('    {}\t|      {} \t| {}'.format(k, int(np.sum(errors)), np.argwhere(errors).T[0]))
```

```python
# Show hard-to-classify bears

distinct_errors = set(errors_list)

for error_index in distinct_errors:
    if errors_list.count(error_index) > 0:
        # Display the corresponding test image in grayscale
        plt.imshow(test_images[error_index], cmap='gray')
        plt.show()
```

# Part 4: Parametric classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
from sklearn.metrics import classification_report, confusion_matrix
# Flatten test images into 1D arrays
flattened_test_images = np.array(test_images).reshape(len(test_images), -1)

# Initialize models
log_reg_model = LogisticRegression(max_iter=5000)
lda_classifier = LinearDiscriminantAnalysis()
qda_classifier = QuadraticDiscriminantAnalysis()

# Train models on training data
log_reg_model.fit(A, y_train)
lda_classifier.fit(A, y_train)
qda_classifier.fit(A, y_train)

# Predict using Logistic Regression
log_reg_predictions = log_reg_model.predict(flattened_test_images)
print("\n--- Logistic Regression ---")
print("Classification Report:\n", classification_report(y_test, log_reg_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_test, log_reg_predictions))

# Predict using LDA
lda_predictions = lda_classifier.predict(flattened_test_images)
print("\n--- Linear Discriminant Analysis ---")
print("Classification Report:\n", classification_report(y_test, lda_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_test, lda_predictions))

# Predict using QDA
qda_predictions = qda_classifier.predict(flattened_test_images)
print("\n--- Quadratic Discriminant Analysis ---")
print("Classification Report:\n", classification_report(y_test, qda_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_test, qda_predictions))
```

```python
# Construct the bear mask:
# Show absolute value of logistic-regression coefficients for each pixel
plt.imshow(np.abs((log_reg_model.coef_).reshape((128,128))))
plt.colorbar()
```

# Part 5: Robustness to additive white noise

### PART 5.

Introducing noise to the images typically reduces test accuracy because it can obscure critical features or patterns that the model relies on for making predictions. Setting the add noise parameter to True decreases the accuracy of Logistic Regression and LDA, but interestingly, it increases the accuracy of QDA. This might be because QDA tends to overfit on clean training data, and the noise helps the data to align better with QDA's more flexible decision boundaries.

When noise is added, the features highlighted in the eigenbears become less pronounced compared to the noise-free version, making the bear mask appear less realistic.

Gaussian noise, which follows a normal distribution and is added to each pixel, might be handled differently compared to randomly additive noise. With Gaussian noise, it is possible that models like logistic regression can form a decision boundary that accommodates the slightly varied data, potentially enhancing its generalization to unseen data.
